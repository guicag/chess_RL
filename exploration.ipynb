```python
import matplotlib.pyplot as plt
import agents
import train
```

```python
agent_q_learning = agents.QLearningAgent()
q_values_agent_q_learning, _, _, rewards_agent_q_learning = train.train_agent(agent_q_learning, episodes=1000)
```

```python
agent_ruql = agents.RepeatedUpdateQLearningAgent()
q_values_agent_ruql, _, _, rewards_agent_ruql = train.train_agent(agent_ruql, episodes=1000)
```

```python
# Create a new figure
plt.figure(figsize=(15, 5))

# Plot the Q-values
plt.plot(q_values_agent_q_learning, label="QL")
plt.plot(q_values_agent_ruql, label="RUQL")

plt.legend()

# Add a title and axis labels
plt.title('Q-Values Over Time')
plt.xlabel('Episode')
plt.ylabel('Q-Value')


# Show the plot
plt.show()
```

```python
# Create a new figure
plt.figure(figsize=(15, 5))

# Plot the rewards
plt.plot(rewards_agent_q_learning, label="QL")
plt.plot(rewards_agent_ruql, label="RUQL")

# Add a legend
plt.legend()

# Add a title and axis labels
plt.title('Cumulative rewards Over Time')
plt.xlabel('Episode')
plt.ylabel('Reward')


# Show the plot
plt.show()
```
